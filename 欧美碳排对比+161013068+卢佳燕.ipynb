# 随机数据bar图

```python

from pyecharts.faker import Faker
from pyecharts import options as opts
from pyecharts.charts import Bar


def bar_base() -> Bar:
    工厂函数 = (
        Bar()
        .add_xaxis(Faker.choose())
        .add_yaxis("商家A", Faker.values())
        .add_yaxis("商家B", Faker.values())
        .set_global_opts(title_opts=opts.TitleOpts(title="Bar-基本示例", subtitle="我是副标题"))
    )
    return 工厂函数
```

```python
柱状图=bar_base()
柱状图.render_notebook()
```

# Pyecharts可视化美国和欧洲历史CO2排放量对比

```python
import pandas as pd
df =pd.read_csv("oecd_carbon_emissions_data.csv",index_col="Country")
df
```

```python
x轴 =[int(x)for x in df.columns.values[-5:]]
x轴
```

```python
x轴2=[str(x) for x in x轴]#折线图x轴转为字符串列表数据
x轴2
```

```python
美国 =list(df.loc["United States"].values)[-4:]
美国
```

```python
欧洲 =list(df.loc["European Union (28 countries)"].values)[-4:]
欧洲
```

# 柱状图

```python

from pyecharts import options as opts
from pyecharts.charts import Bar


def bar_base() -> Bar:
    a = (
        Bar()
        .add_xaxis([int(x)for x in df.columns.values[-5:]])
        .add_yaxis("美国", list(df.loc["United States"].values)[-5:])
        .add_yaxis("欧洲", list(df.loc["European Union (28 countries)"].values)[-5:])
        .set_global_opts(title_opts=opts.TitleOpts(title="近5年欧美国家CO2排放对比",subtitle="2013-2017年数据"))
    )
       
    return a
```

```python
a=bar_base()
a.render_notebook()
```
