```python
from pyecharts import options as opts
from pyecharts.charts import Geo
from pyecharts.globals import ChartType, SymbolType


def geo_new() -> Geo:
    c = (
        Geo()
        .add_schema(maptype="china")
        .add("用户评价", [['广东', 87], ['北京', 50], ['上海', 29],])
        .set_series_opts(label_opts=opts.LabelOpts(is_show=False))
        .set_global_opts(
            visualmap_opts=opts.VisualMapOpts(),
            title_opts=opts.TitleOpts(title="用户评价-基本示例"),
        )
    )
    return c
```

```python
from pyecharts import options as opts
from pyecharts.charts import Bar


def bar_base() -> Bar:
    a = (
        Bar()
        .add_xaxis([int(x)for x in df.columns.values[0,5]])
        .add_yaxis("北京", list(df.loc["北京"].values)[0,5])
        .add_yaxis("云南", list(df.loc["云南)"].values)[0,5])
        .set_global_opts(title_opts=opts.TitleOpts(title="近5年北京与云南氮气排放对比",subtitle="20113-2017年数据"))
    )
       
    return a
```

```python
a=bar_base()
a.render_notebook()
```

```python
# 元组的列表
分省氮气排 = list(zip(list(df.province),list(df.emissions)))
print(分省氮气排)
```

```python
from pyecharts import options as opts
from pyecharts.charts import Geo
from pyecharts.globals import ChartType, SymbolType

def geo_氮气排() -> Geo:
    c = (
        Geo()
        .add_schema(maptype="china")
        .add("氮气排(Mt)", 分省氮气排)
        .set_series_opts(label_opts=opts.LabelOpts(is_show=False))
        .set_global_opts(
            visualmap_opts=opts.VisualMapOpts(),
            title_opts=opts.TitleOpts(title="中国分省氮气排数据"),
        )
    )
    return c
```

```python
分省氮气排地理图 = geo_氮气排()
分省氮气排地理图.render_notebook()
```

数据故事：可以看出我国东南部的氮气排放量比较大
原因：有可能是东南部地区经济比较发达，汽车尾气、工厂废气排放量大所致。
