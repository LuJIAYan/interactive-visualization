```python
# 自己挑函數 apply 
"""
* 10.3 apply：一般性的“拆分－应用－合并”
"""
def baseline(x):
    return (x-100)

df.groupby('地区名').apply(baseline)
```

```python
# 禁止分组键
# 从上面的例子中可以看出，分组键会跟原始对象的索引共同构成结果对象中的层次化索引。将group_keys=False传入groupby即可禁止该效果：

df.groupby('地区名', group_keys=False).apply(baseline)
```

### 4A. 转码为名，给人读
* 要会取用并改写 索引名称/內容 及 栏位索引名称/內容
* 要会使用字典做批次转换

```python
# 处理dp前先观察
print("数据框大小:", dp.shape)
print("索引名称/内容:", dp.index)
print("栏位索引名称/内容:", dp.columns)
dp.head()
```

```python
'''✂———————————————————————————
注意 上栏结果
1. 索引名称/内容, 有2层多重索引，先是reg，再是zb，都有信息量，不能简单简化
2. 栏位索引名称/内容, 有2层多重索引，先是None([['data']])，再是sj，只有一个sj有信息量，可以简单简化
'''

# 使用 .reset_index()把 有2层多重索引 去索引化转成内容
# dp_simp = dp.reset_index().copy()   # 不使用参数
dp_simp = dp.reset_index(col_level=1, col_fill='ind').copy()  # 使用参数

print("数据框大小:", dp_simp.shape)
print("索引名称/内容:", dp_simp.index)
print("栏位索引名称/内容:", dp_simp.columns)
```

```python
# 使用 .droplevel()降多重索引為簡單
dp_simp.columns = dp_simp.columns.droplevel()

print("数据框大小:", dp_simp.shape)
print("索引名称/内容:", dp_simp.index)
print("栏位索引名称/内容:", dp_simp.columns)
```

```python
dp_simp
```

```python
# 转码为名，给人读
# 使用最准备好的dict_misc
print (str(dict_misc)[0:3000]) # 观察 
```

```python
# dict_misc['wd'] 有 'zb' 及 'sj' 的中文
dp_simp.columns = [dict_misc['wd'].get(x,x) for x in dp_simp.columns] # 栏位名称改用中文
print("索引名称/内容:", dp_simp.index)
print("栏位索引名称/内容:", dp_simp.columns)
```

```python
dp_simp
```

```python
# dict_misc['reg'] 有 指标 reg 代码的中文名称  注意 str()转换型别
dp_simp['地区名'] = [dict_misc['reg'].get(str(x),x) for x in dp_simp.地区] # 栏位名称改用中文
dp_simp
```

```python
[dict_misc['zb'].get(x,x).replace("增加值指数(上年=100)","") for x in dp_simp.columns]
```
