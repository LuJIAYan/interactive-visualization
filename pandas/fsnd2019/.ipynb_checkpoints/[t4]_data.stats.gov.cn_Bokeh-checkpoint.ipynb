### 3. 分组数据：重塑reshape
* [根据索引级别分组](https://nbviewer.jupyter.org/github/LearnXu/pydata-notebook/blob/master/Chapter-10/10.1%20GroupBy%20Mechanics%EF%BC%88%E5%88%86%E7%BB%84%E6%9C%BA%E5%88%B6%EF%BC%89.ipynb)
* [stack和unstack的形象理解](https://blog.csdn.net/anshuai_aw1/article/details/82830916)

```python
# 使用reset_index()取消索引
# 使用reset_index()重建索引

df_i = df_s.reset_index().set_index(['zb','reg','sj'])
df_i
```

```python
# unstack -1 层，注意columns 变化
# ['zb','reg',**'sj'**] 
df_i.unstack(level=-1)
```

```python
# unstack 1 层，注意columns 变化
# ['zb',**'reg'**,'sj']
df_i.unstack(level=1)
```

```python
# unstack 0 层，注意columns 变化
# [**'zb'**,'reg','sj']
df_i.unstack(level=0)
```

### 3. 分组数据：统计

## 7. 线性回归
* statsmodels
* OLS model

```python
title = "软件业 vs 科学研究业" 
```

```python
# 线性回归

import statsmodels.formula.api as sm   # http://www.statsmodels.org/devel/examples/notebooks/generated/ols.html
def cal_ols_resid(df, y, x):
    model = sm.ols(formula='{y} ~ {x}'.format(y=y,x=x), data=df )
    fitted = model.fit()
    return(fitted)
```

```python
cal_ols_resid(dfs, y=y, x=x)
```

```python
from bokeh.plotting import figure, ColumnDataSource, output_notebook, show
from bokeh.models import HoverTool
output_file(title+".html")

TOOLS = "pan,wheel_zoom,zoom_in,zoom_out,box_zoom,undo,redo, \
         reset,tap,save,box_select,poly_select,lasso_select," #crosshair,
#['box_zoom', 'reset']

fig = figure(width=600, height=600, tools=TOOLS, title=title)
    # We're getting data from the given dataframe

#显示title
fig.title.text_font_size = "2rem" #"18pt"
fig.title.text_font_style = "normal"
fig.title.text_font = "Microsoft YaHei" #"SimHei"#, 'Microsoft YaHei'

#显示轴标签
fig.xaxis.axis_label = x
fig.yaxis.axis_label = y
fig.xaxis.axis_label_text_font_size ,fig.yaxis.axis_label_text_font_size  = "1rem","1rem"
fig.xaxis.axis_label_text_font_style,fig.yaxis.axis_label_text_font_style = "normal","normal"
fig.xaxis.axis_label_text_font      ,fig.yaxis.axis_label_text_font       = "Microsoft YaHei","Microsoft YaHei"
fig.xaxis.axis_label_standoff = 10
fig.xaxis.major_label_text_font     ,fig.yaxis.major_label_text_font      = "1.6rem","1.6rem"


#df = df.reset_index() # move index to column.
dict_col = {x:"col{i}".format(i=i) for i,x in enumerate(dfs.columns)}
dict_rev = {"col{i}".format(i=i):x for i,x in enumerate(dfs.columns)}

# source cannot have column names in Chinese or unicode
df_ = dfs.copy()
df_.columns = [dict_col[x] for x in df_.columns]

source = ColumnDataSource(ColumnDataSource.from_df(df_))  
name='main'
fig.scatter(dict_col[x], dict_col[y], source=source, name=name, marker='o')


fitted = cal_ols_resid(dfs, y=y, x=x)

b=fitted.params.Intercept
a=fitted.params[x]

# regression line
#fig.line(x=[df[x][i] for i in fitted.fittedvalues.index],
#         y=[fitted.fittedvalues[i] for i in fitted.fittedvalues.index],
#         color = 'blue', line_width=5, line_alpha=0.5)
fig.line(x=[0, dfs[x].max()*1.05],
         y=[b, a*dfs[x].max()*1.05+b],
         color = 'red', line_width=2,)

hover = HoverTool(names=[name], mode='vline')
cols=None

if cols is None:
    # Display *all* columns in the tooltips
    # ("中文", "@desc")
    hover.tooltips = [(dict_rev[c], u'@{c}'.format(c=c)) for c in df_.columns]
    #debug
    #print(hover.tooltips)
else:
    # Display just the given columns in the tooltips
    hover.tooltips = [(dict_rev[c], u'@{c}'.format(c=c)) for c in [dict_col[x] for x in cols]]
    #hover.tooltips.append(('index', '$index'))
# Finally add/enable the tool
fig.add_tools(hover)
show(fig)    
```
