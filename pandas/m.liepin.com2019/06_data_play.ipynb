### 清数据：用函数

```python
# ## 990-990万 保密 面议
import numpy as np

def clean_salary(dlist):
    dlist_output = []
    for data in dlist:
        if '万' in data:
            datac = data.replace('万','')  #data.strip('万')
            datac = datac.replace('以上','-')  #data.strip('万')
            datac = datac.replace('以下','')  #data.strip('万')
            datao = [int(x) if x !='' else 0 for x in datac.split('-')]
        elif '保密' in data or '面议' in data:
            datao = [np.nan,np.nan]
        dlist_output.append(datao)
        
    return(dlist_output)           
    
```

```python
df_s = pd.DataFrame(clean_salary(list(df00.salary)))
df_s.columns=["下","上"]
df00 = df00.drop(["下","上"], axis=1) #删去上、下列?
df00 = df00.join(df_s)
df00
```

### 清数据 location_company: 细化

```python
[x.split(',') for x in df00['location_company']] #把数据变成列表
```

```python
locs = set()   ##把数据变成集合
for x in df00['location_company']:
    locs = locs.union(x.split(',')) 
locs
```

```python
locs_main = set()
for x in df00['location_company']:
    locs = x.split(',')
    #print ([y.split('-')[0] if '-' in y else y  for y in locs])
    locs_main = locs_main.union(set([y.split('-')[0] if '-' in y else y  for y in locs])) 
locs_main
```

```python
print(locs_main)
print(len(locs_main))
```

```python
df00[['location_company']].tail()
```

```python
df_l = df00[['location_company']].copy()

for loc in locs_main:
    df_l[loc] = [True if loc in x else False for x in df_l['location_company']]
df_l
```

```python
df_l[list(locs_main)]
```

```python
loc_freq = df_l[list(locs_main)].stack().reset_index()
loc_freq
```

```python
loc_freq = df_l[list(locs_main)].stack().reset_index().rename(columns={'level_0': 'job_index', 'level_1': 'loc_main', 0:'occr'})
loc_freq
```
